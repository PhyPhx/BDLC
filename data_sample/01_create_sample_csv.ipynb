```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Create Data Samples") \
    .getOrCreate()
```

```python
#Raw Sample
raw_df = spark.read.parquet("/taxi/raw/2021/*.parquet")

raw_sample = raw_df.limit(1000).toPandas()

raw_sample.to_csv(
    "/home/cluster/BDLC/data_sample/yellow_taxi_raw_sample_1000.csv",
    index=False
)
```

```python
#Clean Sample

clean_df = spark.read.parquet("/taxi/clean")

clean_sample = clean_df.sample(
    withReplacement=False,
    fraction=0.001,
    seed=42
).limit(50000).toPandas()

clean_sample.to_csv(
    "/home/cluster/BDLC/data_sample/yellow_taxi_clean_sample_50000.csv",
    index=False
)
```
