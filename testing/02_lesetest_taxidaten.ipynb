```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = SparkSession.builder \
    .appName("Taxi Data Exploration") \
    .getOrCreate()

df_2021 = spark.read.parquet("/taxi/raw/2021/*.parquet")

df_2021.printSchema()
df_2021.show(5)
record_count = df_2021.count()

print(f"Number of records 2021: {record_count}")
```

```python
spark.stop()
```
